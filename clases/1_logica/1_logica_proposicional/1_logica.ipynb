# Llamada aleatoria a participar

```python
import pandas as pd
import random

# Cargar la lista de estudiantes desde la URL
url = 'https://raw.githubusercontent.com/marco-canas/groups_list/main/fund_aritmetica.csv'
estudiantes = pd.read_csv(url)

# Función para llamar a un estudiante aleatoriamente y registrar su nota de seguimiento
def llamar_estudiante():
    estudiante_aleatorio = random.choice(estudiantes['Nombre'])
    print(f"Llamando a {estudiante_aleatorio}...")
    
    nota_seguimiento = float(input("Ingrese la nota de seguimiento para el estudiante (0-5): "))
    
    # Validar que la nota esté en el rango correcto (0-5)
    if nota_seguimiento < 0 or nota_seguimiento > 5:
        print("La nota debe estar en el rango de 0 a 5.")
        return
    
    # Actualizar la nota de seguimiento en el DataFrame
    index_estudiante = estudiantes[estudiantes['Nombre'] == estudiante_aleatorio].index
    estudiantes.at[index_estudiante, 'Seguimiento'] = nota_seguimiento
    
    # Guardar el DataFrame actualizado en un archivo CSV
    estudiantes.to_csv('calificaciones_fundamentos_aritmetica.csv', index=False)
    
    print(f"Nota de seguimiento registrada para {estudiante_aleatorio}: {nota_seguimiento}")
    
# Llamar a la función para realizar el proceso
llamar_estudiante()
```

### 1. Lógica matemática



La lógica es, en términos generales, la ciencia que investiga, formula y establece los principios del
razonamiento. 

La palabra proviene del término latino lógica, y este, del griego logikè, cuya etimología significa “ciencia del discurso” o “ciencia del pensamiento”.

En su sentido más moderno, la lógica no se ocupa de establecer la veracidad del contenido de una determinada afirmación, sino que se encarga de determinar la coherencia de las afirmaciones, es decir, de la validez de las inferencias.

## La ley de la disyunción

La mamá de Juanito le dice: 

Puedes jugar con el celular cuando y sólo cuando laves los trastes o barras la casa. 
De lo dicho por la mamá se infiere que la única menera que Juenito no salga es cuendo no hace ambas labores, es decir, la disyunción solo es falsa cuendo ambas proposiciones atómicas son falsas:

```python
print(r'$P$' + '\t' + r'$Q$' + '\t' + r'$P \vee Q$')
print('-'*26)
valores_verdad = [True, False]

for i in valores_verdad:
    for j in valores_verdad:
        print(str(i) + '\t' + str(j) + '\t' + str(i or j))
        
```

# Conjunción

```python
print(r'$P$' + '\t' + r'$Q$' + '\t' r'$P \wedge Q$')
print('_'*26)

lista_valores_booleanos = [True, False]

for i in lista_valores_booleanos:
    for j in lista_valores_booleanos:
        print(str(i) + '\t' + str(j) + '\t' + str(i and j))
```

# Tabla de verdad del condicional en Python 

```python
print(r'$P$' + '\t' + r'$Q$' + '\t' r'$P \Rightarrow Q$')
print('_'*26)

lista_valores_booleanos = [True, False]

for i in lista_valores_booleanos:
    for j in lista_valores_booleanos:
        print(str(i) + '\t' + str(j) + '\t' + str(not i or j))
```

$$ (P \to Q) \leftrightarrow (\neg P \vee Q) $$

## Construcción de la primera ley de De Morgan utilizando el contrarrecíproco y la capacidad de hacer inferencias de una premisa o suposición. 

## Si lavas los trastes o barres, entonces te dejo salir. 

Luego de que Juanito no  haya salió se infiere que:

$$ \neg ( L \vee B); $$

es decir, 


$$ no labó los trastes y no barrió la casa.  $$


De donde se deriva la siguiente ley lógica:

$$ \neg(P \vee Q) \ \ \ \equiv\ \ \  (\neg P \wedge \neg Q) $$

        R es suficiente para S. Se entiende como "basta que se dé R para que ocurra S. Se puede concebir la implicación en estos términos como un compromiso en el sentido de que si se da el antecedente, entonces, tiene que darse el consecuente. Ahora, si no se da el antecedente, no hay compromiso y, por tanto, el consecuente puede darse o no. Si $P$ es una proposición, a la proposición $P$ y no $P$ se le llama contradicción. ## Participación aleatoria de los estudiantes del curso

```python
def estudiante_que_va_a_participar(grupo): # el grupo se introduce como una cadena de caracteres 
    import numpy as np 
    import pandas as pd 
    from random import choice
    grupo_df = pd.read_csv(grupo + '.csv', encoding = 'latin1', sep = ';')
    lista_estudiantes = list(grupo_df.email.values)
    if lista_estudiantes != []:
        estudiante = choice(lista_estudiantes)
        lista_estudiantes.remove(estudiante)
        return print(estudiante)
    else:
        print('Todos los estudiantes han participado.\n Muchas gracias y Felicitaciones')
```

```python
estudiante_que_va_a_participar('arit') 
```
