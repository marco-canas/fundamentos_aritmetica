### Componentes Históricos

1. **Orígenes de la Lógica Proposicional**
   - **Aristóteles (384-322 a.C.)**: Fundador de la lógica formal y el silogismo, cuya obra influyó en el desarrollo posterior de la lógica.
   - **Gottfried Wilhelm Leibniz (1646-1716)**: Introdujo la idea de un lenguaje universal para la lógica y la matemática, precursor de la lógica proposicional moderna.



2. **Desarrollo de la Lógica Proposicional**
   - **George Boole (1815-1864)**: Publicó "The Laws of Thought" en 1854, donde formalizó la lógica proposicional en términos algebraicos, dando lugar al álgebra de Boole.
   - **Augustus De Morgan (1806-1871)**: Desarrolló las leyes de De Morgan que son fundamentales para la lógica proposicional y la teoría de conjuntos.



3. **Lógica Moderna y Formalización**
   - **Bertrand Russell (1872-1970) y Alfred North Whitehead (1861-1947)**: Su obra "Principia Mathematica" (1910-1913) formalizó la lógica y las bases de las matemáticas, avanzando la lógica proposicional y de predicados.
   - **Kurt Gödel (1906-1978)**: Con sus teoremas de incompletitud, mostró las limitaciones de los sistemas formales, influenciando la comprensión de la lógica y la matemáticas.

### Componentes Epistemológicos

1. **Fundamentos de la Lógica Proposicional**
   - **Definición de Tautología**: Una tautología es una proposición compuesta que es verdadera en todos los posibles valores de verdad de sus proposiciones simples. Es un concepto clave en la lógica proposicional que asegura que ciertas proposiciones siempre serán verdaderas, independientemente de los valores de verdad de las proposiciones individuales.

2. **Interpretaciones y Aplicaciones**
   - **Interpretaciones Verbales**: Las tautologías pueden interpretarse de diversas maneras, tales como "una proposición es verdadera sin importar el valor de las variables", lo cual refuerza la comprensión de la lógica como un lenguaje universal para describir y analizar el razonamiento.

3. **Lenguaje de la Lógica Matemática**
   - **Formalización**: La formalización matemática de la lógica permite expresar proposiciones y tautologías en un lenguaje preciso que es independiente del lenguaje natural, facilitando la comunicación científica y la verificación de la validez de las proposiciones.
   - **Uso de Herramientas Computacionales**: El uso de herramientas como Python para verificar tautologías resalta la importancia de la lógica en el desarrollo de software y en la ciencia de datos, y muestra cómo la lógica matemática se integra con la tecnología moderna para resolver problemas complejos.

4. **Métodos de Verificación**
   - **Métodos Algebraicos**: El uso del álgebra de Boole y otros métodos algebraicos para verificar tautologías y simplificar proposiciones.
   - **Métodos Computacionales**: La aplicación de algoritmos en Python para comprobar la validez de proposiciones lógicas y tautologías mediante la evaluación exhaustiva de todas las combinaciones posibles de valores de verdad.

Estos componentes pueden ayudar a contextualizar el tema de las tautologías en la historia de la lógica y la matemática, así como en el desarrollo de habilidades prácticas y teóricas en la lógica proposicional.

       - Ejemplo: Verificación de una tautología simple usando Python.
     - **Código de Ejemplo:**


```python

from sympy import symbols, Implies, Not, Or, simplify, And
from sympy.logic.boolalg import truth_table

# Definir variables
p, q = symbols('p q')

# Definir proposiciones
formula = And(p, Not(p))

# Crear tabla de verdad
tt = list(truth_table(formula, [p, q]))

# Verificar si la fórmula es una tautología
is_tautology = all(output == True for _, output in tt)

print(f"Formula: {formula}")
print(f"Simplified Formula: {simplify(formula)}")
print(f"Is Tautology: {is_tautology}")
```


   - **Actividad 2: Construcción y Verificación de Tautologías (20 minutos)**
     - **Instrucciones:**
       - Los estudiantes trabajarán en pares para construir sus propias proposiciones lógicas y verificar si son tautologías usando Python.
       - Cada pareja creará al menos tres proposiciones diferentes y determinará si son tautologías.


```python

from sympy import symbols, Implies, Not, Or, simplify, And, Equivalent
from sympy.logic.boolalg import truth_table

# Definir variables
p, q = symbols('p q')

# Definir proposiciones
formula = Equivalent(Not(And(p, q)), Or(Not(p), Not(q)))

# Crear tabla de verdad
tt = list(truth_table(formula, [p, q]))

# Verificar si la fórmula es una tautología
is_tautology = all(output == True for _, output in tt)

print(f"Formula: {formula}")
print(f"Simplified Formula: {simplify(formula)}")
print(f"Is Tautology: {is_tautology}")
```

1. Determina el valor de verdad de la siguiente proposición compuesta elaborando a mano la tabla de verdad de la proposición y verificando con SymPy: 
   $$\neg(p \land q) \iff (\neg p \lor \neg q)$$

2. Determina el valor de verdad de la siguiente proposición compuesta elaborando a mano la tabla de verdad de la proposición y verificando con SymPy: 
   $$p \lor (q \land r) \iff (p \lor q) \land (p \lor r)$$

3. Determina el valor de verdad de la siguiente proposición compuesta elaborando a mano la tabla de verdad de la proposición y verificando con SymPy: 
   $$(p \lor q) \land (\neg p \lor \neg q) \iff \neg(p \iff q)$$

4. Determina el valor de verdad de la siguiente proposición compuesta elaborando a mano la tabla de verdad de la proposición y verificando con SymPy: 
   $$\neg(p \lor \neg q) \iff (p \land q)$$

5. Determina el valor de verdad de la siguiente proposición compuesta elaborando a mano la tabla de verdad de la proposición y verificando con SymPy: 
   $$p \land (q \lor r) \iff (p \land q) \lor (p \land r)$$

6. Determina el valor de verdad de la siguiente proposición compuesta elaborando a mano la tabla de verdad de la proposición y verificando con SymPy: 
   $$(p \land q) \lor (\neg p \land \neg q) \iff \neg(p \iff q)$$

7. Determina el valor de verdad de la siguiente proposición compuesta elaborando a mano la tabla de verdad de la proposición y verificando con SymPy: 
   $$\neg(p \land (q \lor r)) \iff (\neg p \lor (\neg q \land \neg r))$$

8. Determina el valor de verdad de la siguiente proposición compuesta elaborando a mano la tabla de verdad de la proposición y verificando con SymPy: 
   $$(p \lor q) \land (\neg p \lor r) \iff (q \land \neg r) \lor (p \land \neg q)$$

9. Determina el valor de verdad de la siguiente proposición compuesta elaborando a mano la tabla de verdad de la proposición y verificando con SymPy: 
   $$\neg(p \lor \neg q) \land (p \lor q) \iff \neg(p \iff q)$$

10. Determina el valor de verdad de la siguiente proposición compuesta elaborando a mano la tabla de verdad de la proposición y verificando con SymPy: 
    $$(p \lor q) \land (q \lor r) \iff p \lor (q \land r)$$

11. Determina el valor de verdad de la siguiente proposición compuesta elaborando a mano la tabla de verdad de la proposición y verificando con SymPy: 
    $$\neg(p \land \neg q) \iff (\neg p \lor q)$$

12. Determina el valor de verdad de la siguiente proposición compuesta elaborando a mano la tabla de verdad de la proposición y verificando con SymPy: 
    $$(p \land q) \lor (p \land \neg q) \iff p$$

13. Determina el valor de verdad de la siguiente proposición compuesta elaborando a mano la tabla de verdad de la proposición y verificando con SymPy: 
    $$\neg(p \lor \neg q) \lor (\neg p \land q) \iff \neg(p \iff q)$$

14. Determina el valor de verdad de la siguiente proposición compuesta elaborando a mano la tabla de verdad de la proposición y verificando con SymPy: 
    $$(p \land \neg q) \lor (\neg p \land q) \iff p \iff q$$

15. Determina el valor de verdad de la siguiente proposición compuesta elaborando a mano la tabla de verdad de la proposición y verificando con SymPy: 
    $$\neg(p \land (q \lor r)) \iff (\neg p \lor \neg q) \land (\neg p \lor \neg r)$$

16. Determina el valor de verdad de la siguiente proposición compuesta elaborando a mano la tabla de verdad de la proposición y verificando con SymPy: 
    $$(p \land (q \lor r)) \lor (\neg p \land s) \iff (p \lor \neg p) \land (q \lor r) \lor (\neg p \land s)$$

17. Determina el valor de verdad de la siguiente proposición compuesta elaborando a mano la tabla de verdad de la proposición y verificando con SymPy: 
    $$(p \lor \neg q) \land (\neg p \lor q) \iff \neg(p \iff \neg q)$$

18. Determina el valor de verdad de la siguiente proposición compuesta elaborando a mano la tabla de verdad de la proposición y verificando con SymPy: 
    $$\neg(p \lor (q \land \neg r)) \iff (\neg p \land (\neg q \lor r))$$

19. Determina el valor de verdad de la siguiente proposición compuesta elaborando a mano la tabla de verdad de la proposición y verificando con SymPy: 
    $$(p \land \neg q) \lor (\neg p \land \neg r) \iff \neg(p \lor q \lor r)$$

20. Determina el valor de verdad de la siguiente proposición compuesta elaborando a mano la tabla de verdad de la proposición y verificando con SymPy: 
    $$(p \lor q) \land (\neg p \lor \neg q) \iff \neg(p \land q)$$

```python
# Tarea 1 de lectura de ejemplos y completación de procedimientos
# asignación de tarea para realizar antes de clase
# Consiste en leer y completar un ejemplo de texto guía

import pandas as pd
import numpy as np
import requests
np.random.seed(1213)  # establecer semilla aleatoria para hacer los resultados reproducibles



grupo_estudiantes = 'fundamentos_de_la_aritmetica'
nombre_texto_guia = 'Lina_Grajales'
pagina_y_seccion = 'Desde_la_página_15_para_los_ejemplos_Parte_1'
numero_ejemplos_de_la_seccion = 27
numero_de_ejemplos_asignados = 10


# Descargar la lista de  estudiantes
students_url = "https://raw.githubusercontent.com/marco-canas/groups_list/main/" + grupo_estudiantes + ".csv"
examples = list(range(1, numero_ejemplos_de_la_seccion))  

students_data = pd.read_csv(students_url)

# Crear una lista para almacenar las asignaciones
assignments = []

# Asignar ejercicios aleatorios a cada estudiante
for index, student in students_data.iterrows():
    # Seleccionar 5 ejercicios aleatorios sin repetición
    random_exercises = np.random.choice(examples, numero_de_ejemplos_asignados, replace = False)+ 1 
    assignments.append([students_data["Nombre"][index], *random_exercises])  # Agregar asignación a la lista

# Crear un DataFrame con las asignaciones
columns = ["Student"] + [ "Example" + str(i+1) for i in range(numero_de_ejemplos_asignados)]
assignment_df = pd.DataFrame(assignments, columns=columns)

# Guardar el DataFrame en un archivo CSV
assignment_csv = "tarea" + "_" + nombre_texto_guia + "_"+ pagina_y_seccion + ".csv"
assignment_df.to_csv(assignment_csv, index=False)

print("""Tarea 1 de lectura de ejemplos y completación de procedimientos: 
     Si faltan los enunciados de los problemas, no se valora la tarea
      Si faltan de la conclusión verbales, no se valora la tarea 
      """, assignment_csv)
assignment_df
```
